```python
# Convert from JSON to JSONL for training efficiency
input_file_name = '../data/automotive_faults.json'
output_file_name = '../data/automotive_faults.jsonl'

with open(input_file_name, 'r') as f_in:
    data = json.load(f_in)

with open(output_file_name, 'w') as f_out:
    for item in data:
        json.dump(item, f_out)
        f_out.write('\n')
```
